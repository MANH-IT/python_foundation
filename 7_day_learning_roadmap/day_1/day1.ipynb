## 6. Kiểu dữ liệu (Data types)

#### Kiểu dữ liệu cơ bản trong Python: int, float, str, bool

```python
print(type(x))  # <class 'int'>
print(type(y))  # <class 'float'>
print(type(z))  # <class 'str'>
print(type(t))  # <class 'bool'>
```

## 7.  Ép kiểu dữ liệu (Type casting)

#### Dùng các hàm: int(), float(), str(), bool()

```python
x_str = str(x)            # 10 -> '10'
y_int = int(y)            # 5.5 -> 5
flag_int = int(t)         # True -> 1, False -> 0

print("x_str:", x_str, type(x_str))
print("y_int:", y_int, type(y_int))
print("flag_int:", flag_int)
```

```python
s = "123"
s_num = int(s)  # '123' -> 123
print(s_num)  # Kết quả: 123
```

## 8. Toán tử

#### + : cộng
#### - : trừ
#### * : nhân
#### / : chia thực
#### // : chia lấy phần nguyên
#### % : chia lấy dư
#### ** : lũy thừa

```python
a = 7
b = 3

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
```

```python
print("5 ** 2 =", 5**2)  # 25
print("10 % 4 =", 10 % 4)  # 2
```

## 9. Cấu trúc rẽ nhánh (if, elif, else)
## Dùng để kiểm tra điều kiện và thực hiện các hành động khác nhau

## Cú pháp:
##     if (điều_kiện):
##     thực hiện khi điều_kiện đúng
## elif (điều_kiện_khác):
##     thực hiện khi điều_kiện_khác đúng
## else:
##     thực hiện khi tất cả điều kiện đều sai

```python
so = int(input("Nhập một số: "))
if so > 0:
    print("Số dương")
elif so == 0:
    print("Số 0")
else:
    print("Số âm")
```

```python
diem = float(input("Nhập điểm: "))
if diem >= 8.5:
    print("Giỏi")
elif diem >= 6.5:
    print("Khá")
elif diem >= 5.0:
    print("Trung bình")
else:
    print("Yếu")
```

```python
tuoi = int(input("Nhập tuổi: "))
if tuoi >= 18:
    print("Bạn đã đủ tuổi trưởng thành")
else:
    print("Bạn chưa đủ tuổi")
```

```python
chieu_cao = float(input("Nhập chiều cao (m): "))
can_nang = float(input("Nhập cân nặng (kg): "))
bmi = can_nang / (chieu_cao ** 2)

print("Chỉ số BMI của bạn là:", round(bmi, 2))

if bmi < 18.5:
    print("Bạn gầy.")
elif 18.5 <= bmi < 24.9:
    print("Bạn có cân nặng bình thường.")
elif 25 <= bmi < 29.9:
    print("Bạn bị thừa cân.")
else:
    print("Bạn bị béo phì.")
```

#### 10.2. Tính chu vi và diện tích hình chữ nhật

```python
chieu_dai = float(input("Nhập chiều dài: "))
chieu_rong = float(input("Nhập chiều rộng: "))
cv = 2 * (chieu_dai + chieu_rong)
di = chieu_dai * chieu_rong
print("Chu vi:", cv)
print("Diện tích:", di)
```

#### 10.5. Chuyển đổi độ C sang độ F và K

```python
c = float(input("Nhập nhiệt độ (°C): "))
f = (c * 9/5) + 32
k = c + 273.15
print("Nhiệt độ F:", f, "°F")
print("Nhiệt độ K:", k, "K")
```

#### 10.6. Nhập một số và kiểm tra số đó có chia hết cho 2 và 3 hay không

```python
n = int(input("Nhập một số nguyên: "))
if n % 2 == 0 and n % 3 == 0:
    print("Chia hết cho cả 2 và 3")
elif n % 2 == 0:
    print("Chỉ chia hết cho 2")
elif n % 3 == 0:
    print("Chỉ chia hết cho 3")
else:
    print("Không chia hết cho 2 hoặc 3")
```

#### 10.7. Nhập 3 số và tìm số lớn nhất

```python
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
c = float(input("Nhập số c: "))
max_val = a
if b > max_val:
    max_val = b
if c > max_val:
    max_val = c
print("Số lớn nhất là:", max_val)
```

#### 10.8. Kiểm tra năm nhập vào có phải năm nhuận không

```python
nam = int(input("Nhập năm: "))
if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
    print("Đây là năm nhuận")
else:
    print("Đây không phải là năm nhuận")
```
